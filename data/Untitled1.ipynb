```python
import json
f = open('train.json')
train_data = json.load(f)
```

```python
f = open('test_sim.json')
test_data = json.load(f)
```

```python
train_target = set()
for k,v in train_data.items():
    train_target |= set(v)

test_target = set()
for line in test_data:
    test_target.add(line['full-text'])
```

```python
train_target.intersection(test_target)
```
